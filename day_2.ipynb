```
def draw_bar_plots(feat_group, feat_agg='price_value', agg_func=[np.mean, np.median, np.size], feat_sort='mean', 
                   size=50):
    return (
        df.groupby(feat_group)[feat_agg]
        .agg(agg_func)
        .sort_values(ascending=False, by=feat_sort)
        .head(size)
        .plot(kind='bar', figsize=(18, 5), subplots=True)
    )


draw_bar_plots('param_marka-pojazdu')
```

```
draw_bar_plots('param_kraj-pochodzenia', feat_sort='size')
```
